## Diagnostic Tests - Validating Assumptions

```python
# Linearity Check
plt.figure(figsize=(8, 5))
plt.scatter(reg_model6.fittedvalues, residuals, alpha=0.5)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel("Fitted Values")
plt.ylabel("Residuals")
plt.title("Linearity Check: Residuals vs Fitted Values")
plt.show()
```

### Observation
- **The residuals appear to be randomly scattered around zero, indicating that the linearity assumption holds.**


```python
# Homoscedasticity (Breusch-Pagan Test) 
from statsmodels.stats.diagnostic import het_breuschpagan
bp_test = het_breuschpagan(residuals, x_train_const)
print("\nBreusch-Pagan Test (Homoscedasticity):")
print(f"LM Statistic: {bp_test[0]}, p-value: {bp_test[1]}")
```

### Obseravtion
- **Since the p-value (0.633) is much greater than the 0.05 significance level, we fail to reject the null hypothesis.Therefore, there is no significant difference in residual variance, and the assumption of homoscedasticity holds**

```python
# Normality Test
plt.figure(figsize=(8, 5))
sns.histplot(residuals, bins=10, kde=True)
plt.xlabel("Residuals")
plt.title("Histogram of Residuals (Normality Check)")
plt.show() 
```

```python
# Q-Q Plot
import scipy.stats as stats
plt.figure(figsize=(8, 5))
stats.probplot(residuals, dist="norm", plot=plt)
plt.title("Q-Q Plot for Residual Normality")
plt.show()
```

```python
# Shapiro-Wilk test for normality
shapiro_test = shapiro(residuals) 
print("\nShapiro-Wilk Test (Normality of Residuals):")
print(f"Statistic: {shapiro_test[0]}, p-value: {shapiro_test[1]}")
```

##### since p-value = 0.0291 < 0.05. There is statistical evidence that the residuals are not perfectly normally distributed.

```python
# Independence of Errors (Durbin-Watson Test) 
from statsmodels.stats.stattools import durbin_watson
dw_test = durbin_watson(residuals)
print("\nDurbin-Watson Test (Independence of Errors):")
print(f"Durbin-Watson Statistic: {dw_test}")
```

##### Approx 2, no autocorrelation in residuals

## Make Predictions using test data

```python
x_test_const = sm.add_constant(x_test)
y_pred_train = reg_model6.predict(x_train_const)
y_pred_test = reg_model6.predict(x_test_const)

test_results_df = pd.DataFrame({'Actual Price': y_test, 'Predicted Price': y_pred_test})

test_results_file = r"test_Fitted_Results.csv"
test_results_df.to_csv(test_results_file, index=False)
test_results_df
```

```python
# Calculate performance metrics
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
rmse_train = np.sqrt(mean_squared_error(y_train, y_pred_train))
rmse_test = np.sqrt(mean_squared_error(y_test, y_pred_test))
mae_train = mean_absolute_error(y_train, y_pred_train)
mae_test = mean_absolute_error(y_test, y_pred_test)
```

```python
# Print performance metrics
print("\nPerformance Metrics:")
print(f"Train RMSE: {rmse_train:.4f}")
print(f"Test RMSE: {rmse_test:.4f}")
print(f"Train MAE: {mae_train:.4f}")
print(f"Test MAE: {mae_test:.4f}")
print(f"R-squared: {reg_model6.rsquared:.4f}")
print(f"Adjusted R-squared: {reg_model6.rsquared_adj:.4f}")
```


- **Root Mean Squared Error (RMSE)**
The RMSE indicates the average magnitude of the error. On the test set, the model makes predictions that are, on average, **0.50 units** away from the actual satisfaction values. The slightly lower test RMSE compared to train RMSE suggests that the model generalizes well and is not overfitting.
- **Mean Absolute Error (MAE)**
MAE measures the average absolute error. A test MAE of **0.4000** implies that, on average, the prediction error is less than half a point in the satisfaction score. Like RMSE, MAE also confirms consistent performance across train and test sets.
- **Coefficient of Determination (R-squared)**
This value indicates that **72.73%** of the variation in customer satisfaction can be explained by the independent variables in the model. This reflects a strong level of explanatory power.
- **Adjusted R-squared**
Adjusted RÂ² accounts for the number of predictors in the model. The value suggests that **70.60%** of the variance is explained after adjusting for model complexity. The slight drop from RÂ² to Adjusted RÂ² indicates a well-balanced model without unnecessary predictors.
